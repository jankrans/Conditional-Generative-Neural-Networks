```python
import numpy as np
import pandas as pd
import os, sys

sys.path.append('..')
from scipy.ndimage.filters import gaussian_filter1d
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

log_dir = r'../../runs/run9/logger.log'
with open(log_dir, "r") as f:
    data = f.readlines()

loss = [float(row.split('loss:')[1]) for ep,row in enumerate(data)]

plt.figure(figsize=(20,5))
ysmoothed = gaussian_filter1d(loss, sigma=20)
epochs = range(0, len(loss))
plt.plot(epochs, ysmoothed)
plt.axhline(y = 0, color = 'red')
plt.grid(True, alpha=0.4)
array = range(0,100_000,5000)
plt.xticks(array)
plt.show()

```
